```python
# Graphique du nombre de voitures par segment
segment_counts = df['Segment'].value_counts().reset_index()
segment_counts.columns = ['Segment', 'count']

fig = px.bar(
    segment_counts,
    x='Segment',
    y='count',
    title='Répartition des Voitures par Segment',
    labels={'count': 'Nombre de Voitures'}
)
fig.update_traces(marker_color='steelblue')
fig.update_layout(showlegend=False)
fig.show()
```

```python
# Graphique du nombre de voitures par continent
continent_counts = df['Continent'].value_counts().reset_index()
continent_counts.columns = ['Continent', 'count']

fig_continent = px.bar(
    continent_counts,
    x='Continent',
    y='count',
    title='Répartition des Voitures par Continent',
    labels={'count': 'Nombre de Voitures'}
)
fig_continent.update_traces(marker_color='steelblue')
fig_continent.update_layout(showlegend=False)
fig_continent.show()
```

```python
# Graphique du nombre de voitures par tranche d'année (5 ans)
import numpy as np

# Créer des tranches d'années de 5 ans
annee_min = int(df['Année'].min())
annee_max = int(df['Année'].max())
bins = range(annee_min, annee_max + 6, 5)

df['Tranche_Année'] = pd.cut(df['Année'], bins=bins, right=False)

# Compter les voitures par tranche
year_counts = df['Tranche_Année'].value_counts().sort_index().reset_index()
year_counts.columns = ['Tranche_Année', 'count']
year_counts['Tranche_Année'] = year_counts['Tranche_Année'].astype(str)

fig_year = px.bar(
    year_counts,
    x='Tranche_Année',
    y='count',
    title='Répartition des Voitures par Tranche d\'Année (5 ans)',
    labels={'Tranche_Année': 'Tranche d\'Année', 'count': 'Nombre de Voitures'}
)
fig_year.update_traces(marker_color='steelblue')
fig_year.update_layout(showlegend=False, xaxis_tickangle=-45)
fig_year.show()
```

```python
# Graphique du nombre de voitures par tranche de puissance (50 ch, à partir de 150 ch)

# Déterminer la puissance maximale
puissance_max = int(df['Puissance (ch)'].max())

# Créer des tranches de puissance de 50 ch à partir de 150 ch
bins_puissance = list(range(150, puissance_max + 51, 50))

# Créer la colonne avec les tranches de puissance
df['Tranche_Puissance'] = pd.cut(df['Puissance (ch)'], bins=bins_puissance, right=False)

# Compter les voitures par tranche de puissance
power_counts = df['Tranche_Puissance'].value_counts().sort_index().reset_index()
power_counts.columns = ['Tranche_Puissance', 'count']
power_counts['Tranche_Puissance'] = power_counts['Tranche_Puissance'].astype(str)

fig_power = px.bar(
    power_counts,
    x='Tranche_Puissance',
    y='count',
    title='Répartition des Voitures par Tranche de Puissance (50 ch, à partir de 150 ch)',
    labels={'Tranche_Puissance': 'Tranche de Puissance (ch)', 'count': 'Nombre de Voitures'}
)
fig_power.update_traces(marker_color='steelblue')
fig_power.update_layout(showlegend=False, xaxis_tickangle=-45)
fig_power.show()
```

```python
# Graphique du nombre de voitures par tranche de poids (200 kg)

# Déterminer le poids minimal et maximal
poids_min = int(df['Poids (kg)'].min())
poids_max = int(df['Poids (kg)'].max())

# Arrondir le poids min au multiple de 200 inférieur
poids_min_arrondi = (poids_min // 200) * 200

# Créer des tranches de poids de 200 kg
bins_poids = list(range(poids_min_arrondi, poids_max + 201, 200))

# Créer la colonne avec les tranches de poids
df['Tranche_Poids'] = pd.cut(df['Poids (kg)'], bins=bins_poids, right=False)

# Compter les voitures par tranche de poids
weight_counts = df['Tranche_Poids'].value_counts().sort_index().reset_index()
weight_counts.columns = ['Tranche_Poids', 'count']
weight_counts['Tranche_Poids'] = weight_counts['Tranche_Poids'].astype(str)

fig_weight = px.bar(
    weight_counts,
    x='Tranche_Poids',
    y='count',
    title='Répartition des Voitures par Tranche de Poids (200 kg)',
    labels={'Tranche_Poids': 'Tranche de Poids (kg)', 'count': 'Nombre de Voitures'}
)
fig_weight.update_traces(marker_color='steelblue')
fig_weight.update_layout(showlegend=False, xaxis_tickangle=-45)
fig_weight.show()
```

```python
# Graphique du nombre de voitures par tranche de note presse

# Créer des tranches de notes détaillées à partir de 6/10
# De 6 à 8 : tous les 1
# De 8 à 9 : tous les 0.4
# De 9 à 10 : tous les 0.2
bins_note = [6, 7, 8, 8.4, 8.8, 9.0, 9.2, 9.4, 9.6, 9.8, 10.0]

# Créer la colonne avec les tranches de notes
df['Tranche_Note'] = pd.cut(df['Note Presse (/10)'], bins=bins_note, right=True, include_lowest=True)

# Compter les voitures par tranche de note
note_counts = df['Tranche_Note'].value_counts().sort_index().reset_index()
note_counts.columns = ['Tranche_Note', 'count']
note_counts['Tranche_Note'] = note_counts['Tranche_Note'].astype(str)

fig_note = px.bar(
    note_counts,
    x='Tranche_Note',
    y='count',
    title='Répartition des Voitures par Tranche de Note Presse (à partir de 6/10)',
    labels={'Tranche_Note': 'Tranche de Note (/10)', 'count': 'Nombre de Voitures'}
)
fig_note.update_traces(marker_color='steelblue')
fig_note.update_layout(showlegend=False, xaxis_tickangle=-45)
fig_note.show()
```

```python
# Graphique du nombre de voitures par catégorie de volume de coffre

# Créer des catégories avec les volumes en chiffres
def categoriser_coffre(volume):
    if volume < 200:
        return '0-200 L'
    elif volume < 400:
        return '200-400 L'
    else:
        return '400+ L'

df['Catégorie_Coffre'] = df['Volume Coffre (L)'].apply(categoriser_coffre)

# Compter les voitures par catégorie
coffre_counts = df['Catégorie_Coffre'].value_counts().reindex(['0-200 L', '200-400 L', '400+ L']).reset_index()
coffre_counts.columns = ['Catégorie_Coffre', 'count']

fig_coffre = px.bar(
    coffre_counts,
    x='Catégorie_Coffre',
    y='count',
    title='Répartition des Voitures par Volume de Coffre',
    labels={'Catégorie_Coffre': 'Volume de Coffre', 'count': 'Nombre de Voitures'}
)
fig_coffre.update_traces(marker_color='steelblue')
fig_coffre.update_layout(showlegend=False)
fig_coffre.show()
```

```python
# Top 5 des voitures thermiques les plus puissantes

# Filtrer les voitures thermiques (exclure les électriques)
df_thermiques = df[~df['Segment'].str.contains('Électrique', na=False)]

top5_thermiques = df_thermiques.nlargest(5, 'Puissance (ch)')[['Marque', 'Modèle / Variante', 'Puissance (ch)', 'Année', 'Segment']]

print("🏆 Top 5 des Voitures Thermiques les Plus Puissantes\n")
display(top5_thermiques)
```

```python
# Top 5 des véhicules les plus polyvalents : Breaks, SUV et Berlines (volume de coffre + puissance)

# Filtrer les breaks, SUV et berlines
df_polyvalents = df[df['Segment'].str.contains('Break|SUV|Berline', case=False, na=False)]

print(f"Nombre de véhicules trouvés : {len(df_polyvalents)}\n")

if len(df_polyvalents) > 0:
    # Créer un score de polyvalence : normaliser volume et puissance puis les additionner
    df_polyvalents = df_polyvalents.copy()
    df_polyvalents['Score_Polyvalence'] = (
        (df_polyvalents['Volume Coffre (L)'] - df_polyvalents['Volume Coffre (L)'].min()) / 
        (df_polyvalents['Volume Coffre (L)'].max() - df_polyvalents['Volume Coffre (L)'].min()) +
        (df_polyvalents['Puissance (ch)'] - df_polyvalents['Puissance (ch)'].min()) / 
        (df_polyvalents['Puissance (ch)'].max() - df_polyvalents['Puissance (ch)'].min())
    )
    
    top5_polyvalents = df_polyvalents.nlargest(5, 'Score_Polyvalence')[['Marque', 'Modèle / Variante', 'Puissance (ch)', 'Volume Coffre (L)', 'Segment', 'Année']]
    
    print("🏆 Top 5 des Véhicules les Plus Polyvalents (Breaks, SUV, Berlines)\n")
    display(top5_polyvalents)
else:
    print("❌ Aucun véhicule trouvé dans ces catégories.")
```
