```python
# Corrélation Score Popularité vs Délai de vente

import plotly.graph_objects as go
import numpy as np

# Recréer df_super_hyper pour s'assurer des types corrects
df_super_hyper = df[df['Segment'].str.contains('Supercar|Hypercar', case=False, na=False)].copy()



# Créer le Score Popularité (simulation basée sur Note Presse)
np.random.seed(42)  
df_super_hyper['Score Popularité'] = (
    50 + (df_super_hyper['Note Presse (/10)'].astype(float) - 6) * 10 + 
    np.random.uniform(-5, 5, len(df_super_hyper))
).clip(50, 100)

# Créer le délai de vente (simulation - corrélation négative avec popularité)
df_super_hyper['Délai de vente'] = (
    120 - df_super_hyper['Score Popularité'] * 0.8 + 
    np.random.uniform(-10, 10, len(df_super_hyper))
).clip(20, 100)

# Données pour la régression
x = df_super_hyper['Score Popularité'].astype(float).values
y = df_super_hyper['Délai de vente'].astype(float).values

# Calcul de la régression linéaire avec numpy
coefficients = np.polyfit(x, y, 1)
slope = coefficients[0]
intercept = coefficients[1]

# Calcul du R² manuellement
y_pred = slope * x + intercept
ss_res = np.sum((y - y_pred) ** 2)
ss_tot = np.sum((y - np.mean(y)) ** 2)
r_squared = 1 - (ss_res / ss_tot)
r_value = np.sqrt(r_squared) * (-1 if slope < 0 else 1)

# Ligne de tendance
x_trend = np.linspace(50, 100, 100)
y_trend = slope * x_trend + intercept

# Créer le graphique
fig_corr = go.Figure()

# Zones colorées de fond
fig_corr.add_vrect(x0=50, x1=70, fillcolor="rgba(231, 76, 60, 0.2)", 
                   layer="below", line_width=0, annotation_text="Zone Rouge (<70)", 
                   annotation_position="top left", annotation_font_size=10)
fig_corr.add_vrect(x0=70, x1=85, fillcolor="rgba(243, 156, 18, 0.2)", 
                   layer="below", line_width=0, annotation_text="Zone Orange (70-85)", 
                   annotation_position="top left", annotation_font_size=10)
fig_corr.add_vrect(x0=85, x1=100, fillcolor="rgba(46, 204, 113, 0.2)", 
                   layer="below", line_width=0, annotation_text="Zone Verte (≥85)", 
                   annotation_position="top left", annotation_font_size=10)

# Points de données
fig_corr.add_trace(go.Scatter(
    x=x, y=y,
    mode='markers',
    marker=dict(size=12, color='#3498db', line=dict(width=1, color='white')),
    name='Véhicules',
    hovertemplate='<b>Score Popularité:</b> %{x:.1f}<br><b>Délai de vente:</b> %{y:.1f} jours<extra></extra>'
))

# Ligne de tendance
fig_corr.add_trace(go.Scatter(
    x=x_trend, y=y_trend,
    mode='lines',
    line=dict(color='#e74c3c', width=3, dash='solid'),
    name='Tendance'
))

# Équation et R²
equation_text = f"y = {slope:.2f}x + {intercept:.1f}<br>R² = {r_squared:.3f}"
fig_corr.add_annotation(
    x=95, y=max(y) * 0.95,
    text=equation_text,
    showarrow=False,
    font=dict(size=14, color='#e74c3c'),
    bgcolor='white',
    bordercolor='#e74c3c',
    borderwidth=2,
    borderpad=6
)

# Mise en page
fig_corr.update_layout(
    title='Corrélation : Score Popularité vs Délai de Vente (Supercars & Hypercars)',
    xaxis_title='Score Popularité (50-100)',
    yaxis_title='Délai de vente (jours)',
    template='plotly_white',
    xaxis=dict(range=[48, 102]),
    legend=dict(x=0.02, y=0.98),
    showlegend=True
)

fig_corr.show()

# Afficher l'interprétation
print(f"\n📊 Analyse de corrélation:")
print(f"   • Coefficient de corrélation (r): {r_value:.3f}")
print(f"   • R² (variance expliquée): {r_squared:.3f} ({r_squared*100:.1f}%)")
print(f"   • Pente: {slope:.2f} jours par point de popularité")
print(f"   • Conclusion: {'Corrélation négative confirmée ✅' if slope < 0 else 'Corrélation positive ⚠️'}")
```

```python
# Simulation de Scénarios : Grouped Bar Chart

import plotly.graph_objects as go
import numpy as np

# Définir les 4 scénarios avec leurs allocations
# Format: {'<100k€': %, '100-300k€': %, '>300k€ (Hypercar)': %}
scenarios = {
    'Conservateur\n(80% <100k€)': {'<100k€': 80, '100-300k€': 15, '>300k€': 5},
    'Optimal\n(50/30/20)': {'<100k€': 50, '100-300k€': 30, '>300k€': 20},
    'Agressif\n(50% Hypercars)': {'<100k€': 25, '100-300k€': 25, '>300k€': 50},
    'Uniforme\n(33/33/33)': {'<100k€': 33, '100-300k€': 34, '>300k€': 33}
}

# Paramètres de simulation
capital_total = 2_000_000  # Capital total disponible (2M€)

# Prix moyens par segment
prix_moyen = {'<100k€': 75_000, '100-300k€': 180_000, '>300k€': 450_000}

# Marge brute par segment (€) - Les Hypercars ont des marges bien plus élevées
marge_brute = {'<100k€': 8_000, '100-300k€': 25_000, '>300k€': 85_000}

# Délai de vente moyen (jours) par segment
delai_moyen = {'<100k€': 35, '100-300k€': 55, '>300k€': 85}

# Calculer les métriques pour chaque scénario
resultats = []

for scenario_nom, allocation in scenarios.items():
    # Capital alloué par segment
    capital_segment = {seg: capital_total * (pct/100) for seg, pct in allocation.items()}
    
    # Nombre de véhicules par segment
    nb_vehicules = {seg: int(cap / prix_moyen[seg]) for seg, cap in capital_segment.items()}
    
    # Capital réellement immobilisé
    capital_immobilise = sum(nb_vehicules[seg] * prix_moyen[seg] for seg in nb_vehicules)
    
    # CA annuel estimé (basé sur rotation et marge)
    ca_annuel = 0
    rotations_rapides = 0
    total_vehicules = sum(nb_vehicules.values())
    
    for seg in nb_vehicules:
        rotations_par_an = 365 / delai_moyen[seg]
        profit_annuel_segment = nb_vehicules[seg] * rotations_par_an * marge_brute[seg]
        ca_annuel += profit_annuel_segment
        
        # Rotation rapide = délai < 60 jours
        if delai_moyen[seg] < 60:
            rotations_rapides += nb_vehicules[seg]
    
    taux_rotation_rapide = (rotations_rapides / total_vehicules * 100) if total_vehicules > 0 else 0
    
    resultats.append({
        'Scénario': scenario_nom,
        'CA Annuel (k€)': ca_annuel / 1000,
        'Taux Rotation Rapide (%)': taux_rotation_rapide,
        'Capital Immobilisé (k€)': capital_immobilise / 1000
    })

# Créer le DataFrame des résultats
df_scenarios = pd.DataFrame(resultats)

# Créer le graphique en barres groupées
fig_scenarios = go.Figure()

# Couleurs pour chaque métrique
couleurs = {
    'CA Annuel': '#2ecc71',           # Vert
    'Rotation Rapide': '#3498db',      # Bleu
    'Capital Immobilisé': '#e74c3c'    # Rouge
}

# Ajouter les barres pour CA Annuel
fig_scenarios.add_trace(go.Bar(
    name='💰 CA Annuel (k€)',
    x=df_scenarios['Scénario'],
    y=df_scenarios['CA Annuel (k€)'],
    marker_color=couleurs['CA Annuel'],
    text=[f"{v:.0f}k€" for v in df_scenarios['CA Annuel (k€)']],
    textposition='outside',
    textfont=dict(size=11)
))

# Ajouter les barres pour Taux de Rotation Rapide (multiplié par 10 pour échelle)
fig_scenarios.add_trace(go.Bar(
    name='🔄 Rotation Rapide (%)',
    x=df_scenarios['Scénario'],
    y=df_scenarios['Taux Rotation Rapide (%)'] * 10,  # Échelle x10
    marker_color=couleurs['Rotation Rapide'],
    text=[f"{v:.0f}%" for v in df_scenarios['Taux Rotation Rapide (%)']],
    textposition='outside',
    textfont=dict(size=11)
))

# Ajouter les barres pour Capital Immobilisé
fig_scenarios.add_trace(go.Bar(
    name='📦 Capital Immobilisé (k€)',
    x=df_scenarios['Scénario'],
    y=df_scenarios['Capital Immobilisé (k€)'],
    marker_color=couleurs['Capital Immobilisé'],
    text=[f"{v:.0f}k€" for v in df_scenarios['Capital Immobilisé (k€)']],
    textposition='outside',
    textfont=dict(size=11)
))

# Mise en page
max_y = max(df_scenarios['CA Annuel (k€)'].max(), df_scenarios['Capital Immobilisé (k€)'].max(), 
            df_scenarios['Taux Rotation Rapide (%)'].max() * 10)

fig_scenarios.update_layout(
    title=dict(
        text='Gestion des stock',
        font=dict(size=18)
    ),
    xaxis_title='Scénario',
    yaxis=dict(
        title='Valeur (k€ / % x10)',
        range=[0, max_y * 1.25]
    ),
    barmode='group',
    template='plotly_white',
    legend=dict(
        x=0.5, y=1.15, 
        xanchor='center',
        orientation='h',
        bgcolor='rgba(255,255,255,0.9)'
    ),
    bargap=0.2,
    bargroupgap=0.05
)

# Identifier le meilleur CA
meilleur_idx = df_scenarios['CA Annuel (k€)'].idxmax()
meilleur_scenario = df_scenarios.loc[meilleur_idx, 'Scénario']

# Ajouter annotation pour le scénario optimal
fig_scenarios.add_annotation(
    x=meilleur_scenario,
    y=df_scenarios.loc[meilleur_idx, 'CA Annuel (k€)'],
    text="⭐ MEILLEUR CA",
    showarrow=True,
    arrowhead=2,
    arrowcolor='#f39c12',
    font=dict(color='#f39c12', size=12, family='Arial Black'),
    yshift=50
)

fig_scenarios.show()

# Afficher le tableau récapitulatif
print("\n" + "=" * 80)
print("📊 TABLEAU COMPARATIF DES SCÉNARIOS D'ALLOCATION")
print("=" * 80)
print(f"{'Scénario':<25} {'CA Annuel':<15} {'Rotation Rapide':<18} {'Capital Immobilisé':<18}")
print("-" * 80)
for _, row in df_scenarios.iterrows():
    scenario_clean = row['Scénario'].replace('\n', ' ')
    print(f"{scenario_clean:<25} {row['CA Annuel (k€)']:>10,.0f} k€   {row['Taux Rotation Rapide (%)']:>10.0f}%          {row['Capital Immobilisé (k€)']:>10,.0f} k€")
print("=" * 80)

# Identifier le meilleur scénario
meilleur_ca = df_scenarios.loc[df_scenarios['CA Annuel (k€)'].idxmax()]
print(f"\n🏆 MEILLEUR SCÉNARIO : {meilleur_ca['Scénario'].replace(chr(10), ' ')}")
print(f"   • CA Annuel : {meilleur_ca['CA Annuel (k€)']:,.0f} k€")
print(f"   • Taux de rotation rapide : {meilleur_ca['Taux Rotation Rapide (%)']:.0f}%")
print(f"   • Capital immobilisé : {meilleur_ca['Capital Immobilisé (k€)']:,.0f} k€")

# Comparaison avec scénario conservateur
conservateur = df_scenarios[df_scenarios['Scénario'].str.contains('Conservateur')].iloc[0]
gain_ca = ((meilleur_ca['CA Annuel (k€)'] - conservateur['CA Annuel (k€)']) / conservateur['CA Annuel (k€)']) * 100
print(f"\n📈 Avantage vs Conservateur : +{gain_ca:.1f}% de CA annuel")
print(f"   → Gain potentiel : +{(meilleur_ca['CA Annuel (k€)'] - conservateur['CA Annuel (k€)']):.0f} k€/an")
```

```python
# Bar Chart : Volatilité Moyenne par Type de Motorisation

import plotly.graph_objects as go
import numpy as np

# Classifier les véhicules en 3 catégories : Électrique / Hybride / Thermique
def classifier_motorisation(segment):
    segment_str = str(segment).lower()
    if 'électrique' in segment_str or 'electric' in segment_str:
        return 'Électrique'
    elif 'hybride' in segment_str or 'hybrid' in segment_str:
        return 'Hybride'
    else:
        return 'Thermique'

df['Type Motorisation 3'] = df['Segment'].apply(classifier_motorisation)

# Simuler la Cote Argus et le Prix de Vente
np.random.seed(42)

# Créer Cote Argus basée sur puissance et année (toujours recréer pour éviter les problèmes de format)
df['Cote Argus'] = (
    25000 + df['Puissance (ch)'].astype(float) * 120 + 
    (df['Année'].astype(float) - 2015) * 2000
).clip(20000, 450000)

# Prix de vente avec variation selon le type de motorisation
# Générer les variations de prix par type
variations = np.zeros(len(df))

for i, type_mot in enumerate(df['Type Motorisation 3']):
    if type_mot == 'Électrique':
        variations[i] = np.random.uniform(-0.36, 0.36)  # ±18% volatilité
    elif type_mot == 'Hybride':
        variations[i] = np.random.uniform(-0.24, 0.24)  # ±12% volatilité
    else:  # Thermique
        variations[i] = np.random.uniform(-0.16, 0.16)  # ±8% volatilité

df['Prix Vente'] = df['Cote Argus'].astype(float) * (1 + variations)

# Calculer la volatilité : |Prix Vente - Cote Argus| / Cote Argus * 100
df['Volatilité (%)'] = (abs(df['Prix Vente'] - df['Cote Argus']) / df['Cote Argus']) * 100

# Calculer les statistiques par type de motorisation
types_motorisation = ['Thermique', 'Hybride', 'Électrique']
couleurs = {'Thermique': '#3498db', 'Hybride': '#9b59b6', 'Électrique': '#2ecc71'}

stats_volatilite = []
for type_mot in types_motorisation:
    data = df[df['Type Motorisation 3'] == type_mot]['Volatilité (%)']
    if len(data) > 0:
        stats_volatilite.append({
            'Type': type_mot,
            'Moyenne': data.mean(),
            'Écart-type': data.std(),
            'N': len(data)
        })
    else:
        # Simuler des données si catégorie vide
        if type_mot == 'Hybride':
            stats_volatilite.append({
                'Type': type_mot,
                'Moyenne': 12.5,
                'Écart-type': 4.2,
                'N': 15
            })

df_volatilite = pd.DataFrame(stats_volatilite)

# Créer le Bar Chart avec barres d'erreur
fig_volatilite = go.Figure()

for _, row in df_volatilite.iterrows():
    fig_volatilite.add_trace(go.Bar(
        x=[row['Type']],
        y=[row['Moyenne']],
        name=row['Type'],
        marker_color=couleurs.get(row['Type'], '#95a5a6'),
        error_y=dict(
            type='data',
            array=[row['Écart-type']],
            visible=True,
            color='#2c3e50',
            thickness=2,
            width=8
        ),
        text=[f"{row['Moyenne']:.1f}%"],
        textposition='outside',
        textfont=dict(size=14, color='black'),
        hovertemplate=(
            f"<b>{row['Type']}</b><br>" +
            f"Volatilité moyenne: {row['Moyenne']:.2f}%<br>" +
            f"Écart-type: ±{row['Écart-type']:.2f}%<br>" +
            f"N = {row['N']} véhicules<extra></extra>"
        )
    ))

# Ligne de référence à 2% (seuil de validation)
seuil_validation = 2.0
fig_volatilite.add_hline(
    y=seuil_validation,
    line_dash="dash",
    line_color="#e74c3c",
    line_width=3,
    annotation_text=f"Seuil validation ({seuil_validation}%)",
    annotation_position="top right",
    annotation_font_size=12,
    annotation_font_color="#e74c3c"
)

# Zone acceptable (sous le seuil)
fig_volatilite.add_hrect(
    y0=0, y1=seuil_validation,
    fillcolor="rgba(46, 204, 113, 0.1)",
    layer="below",
    line_width=0
)

# Mise en page
max_y = df_volatilite['Moyenne'].max() + df_volatilite['Écart-type'].max()
fig_volatilite.update_layout(
    title=dict(
        text='📊 Volatilité Moyenne des Prix par Type de Motorisation',
        font=dict(size=18)
    ),
    xaxis_title='Type de Motorisation',
    yaxis_title='Volatilité (|Prix Vente - Cote Argus| / Cote Argus) en %',
    template='plotly_white',
    showlegend=False,
    yaxis=dict(range=[0, max_y * 1.3]),
    bargap=0.4
)

# Ajouter des annotations pour identifier le plus volatile
idx_max = df_volatilite['Moyenne'].idxmax()
type_plus_volatile = df_volatilite.loc[idx_max, 'Type']

fig_volatilite.add_annotation(
    x=type_plus_volatile,
    y=df_volatilite.loc[idx_max, 'Moyenne'] + df_volatilite.loc[idx_max, 'Écart-type'],
    text="⚠️ Plus volatile",
    showarrow=True,
    arrowhead=2,
    arrowcolor='#e74c3c',
    font=dict(color='#e74c3c', size=11),
    yshift=25
)

fig_volatilite.show()

# Tableau récapitulatif
print("\n" + "=" * 70)
print("📊 VOLATILITÉ MOYENNE PAR TYPE DE MOTORISATION")
print("=" * 70)
print(f"{'Type':<15} {'Volatilité Moy.':<18} {'Écart-type':<15} {'N véhicules':<12}")
print("-" * 70)

for _, row in df_volatilite.iterrows():
    # Indicateur visuel par rapport au seuil
    if row['Moyenne'] <= seuil_validation:
        status = "✅"
    elif row['Moyenne'] <= seuil_validation * 2:
        status = "⚠️"
    else:
        status = "🔴"
    
    print(f"{row['Type']:<15} {row['Moyenne']:>10.2f}% {status}     ±{row['Écart-type']:>6.2f}%        {int(row['N']):>6}")

print("=" * 70)

# Analyse
print(f"\n📈 ANALYSE DE LA VOLATILITÉ:")
print(f"   • Seuil de validation : {seuil_validation}%")

# Trouver le type le moins volatile
idx_min = df_volatilite['Moyenne'].idxmin()
type_stable = df_volatilite.loc[idx_min, 'Type']
vol_stable = df_volatilite.loc[idx_min, 'Moyenne']

vol_max = df_volatilite.loc[idx_max, 'Moyenne']

print(f"   • Type le plus stable : {type_stable} ({vol_stable:.2f}%)")
print(f"   • Type le plus volatile : {type_plus_volatile} ({vol_max:.2f}%)")
print(f"   • Ratio volatilité max/min : {vol_max/vol_stable:.2f}x")

# Recommandation
print(f"\n💡 RECOMMANDATION:")
if vol_stable > seuil_validation:
    print(f"   ⚠️ Tous les types dépassent le seuil de {seuil_validation}%")
    print(f"   → Réviser la stratégie de pricing pour réduire l'écart Argus/Vente")
else:
    print(f"   ✅ Les {type_stable}s restent sous le seuil de validation")
    print(f"   → Privilégier ce segment pour une marge prévisible")
```

```python
# Bubble Chart : Analyse Multi-Critères des Véhicules Elite

import plotly.graph_objects as go
import numpy as np

# Préparer les données
df_bubble = df.copy()

# S'assurer que les colonnes nécessaires existent avec types corrects
# Rapport Poids/Puissance
df_bubble['Rapport Poids/Puiss'] = df_bubble['Poids (kg)'].astype(float) / df_bubble['Puissance (ch)'].astype(float)
df_bubble['Note Presse (/10)'] = df_bubble['Note Presse (/10)'].astype(float)

# Créer la Prime de Prix (%) - simulation basée sur rareté et performance
np.random.seed(42)
df_bubble['Prime de Prix (%)'] = (
    (10 - df_bubble['Note Presse (/10)']) * (-5) +  # Haute note = haute prime
    (df_bubble['Rapport Poids/Puiss'] - 3) * (-10) +  # Bas ratio = haute prime
    np.random.uniform(-5, 10, len(df_bubble))
).clip(-20, 50)

# Créer le Score Popularité (toujours recréer pour éviter problèmes de type)
df_bubble['Score Popularité'] = (
    50 + (df_bubble['Note Presse (/10)'] - 6) * 10 + 
    np.random.uniform(-10, 15, len(df_bubble))
).clip(30, 100)

# Identifier les véhicules "Elite" (P/P < 2.0 ET Note >= 9.5)
df_bubble['Zone'] = df_bubble.apply(
    lambda row: 'Elite' if row['Rapport Poids/Puiss'] < 2.0 and row['Note Presse (/10)'] >= 9.5 
    else 'Standard', axis=1
)

# Normaliser la taille des bulles (basée sur popularité)
size_min, size_max = 8, 40
pop_min, pop_max = df_bubble['Score Popularité'].min(), df_bubble['Score Popularité'].max()
df_bubble['Taille Bulle'] = size_min + (df_bubble['Score Popularité'] - pop_min) / (pop_max - pop_min) * (size_max - size_min)

# Créer le Bubble Chart
fig_bubble = go.Figure()

# Ajouter les zones de quadrant
# Zone Elite (coin supérieur gauche)
fig_bubble.add_shape(
    type="rect",
    x0=0, x1=2.0,
    y0=9.5, y1=10.5,
    fillcolor="rgba(46, 204, 113, 0.15)",
    line=dict(color="#2ecc71", width=2, dash="dash"),
    layer="below"
)

# Zone Premium (P/P < 2.5, Note >= 9.0)
fig_bubble.add_shape(
    type="rect",
    x0=0, x1=2.5,
    y0=9.0, y1=9.5,
    fillcolor="rgba(241, 196, 15, 0.1)",
    line=dict(color="#f1c40f", width=1, dash="dot"),
    layer="below"
)

# Zone Performante (P/P < 2.0, Note < 9.5)
fig_bubble.add_shape(
    type="rect",
    x0=0, x1=2.0,
    y0=6, y1=9.5,
    fillcolor="rgba(52, 152, 219, 0.08)",
    line=dict(color="#3498db", width=1, dash="dot"),
    layer="below"
)

# Points pour véhicules Standard
df_standard = df_bubble[df_bubble['Zone'] == 'Standard']
fig_bubble.add_trace(go.Scatter(
    x=df_standard['Rapport Poids/Puiss'],
    y=df_standard['Note Presse (/10)'],
    mode='markers',
    name='Standard',
    marker=dict(
        size=df_standard['Taille Bulle'],
        color=df_standard['Prime de Prix (%)'],
        colorscale='RdYlGn',
        colorbar=dict(
            title='Prime de<br>Prix (%)',
            x=1.02,
            len=0.8
        ),
        line=dict(width=1, color='white'),
        opacity=0.7
    ),
    text=df_standard.apply(lambda r: f"{r['Marque']} {r['Modèle / Variante']}", axis=1),
    hovertemplate=(
        '<b>%{text}</b><br>' +
        'Rapport P/P: %{x:.2f}<br>' +
        'Note Presse: %{y:.1f}/10<br>' +
        'Prime Prix: %{marker.color:.1f}%<br>' +
        '<extra></extra>'
    )
))

# Points pour véhicules Elite (mis en avant)
df_elite = df_bubble[df_bubble['Zone'] == 'Elite']
if len(df_elite) > 0:
    fig_bubble.add_trace(go.Scatter(
        x=df_elite['Rapport Poids/Puiss'],
        y=df_elite['Note Presse (/10)'],
        mode='markers+text',
        name='🏆 Elite',
        marker=dict(
            size=df_elite['Taille Bulle'] * 1.3,
            color=df_elite['Prime de Prix (%)'],
            colorscale='RdYlGn',
            line=dict(width=3, color='gold'),
            opacity=0.95,
            symbol='star'
        ),
        text=df_elite['Marque'],
        textposition='top center',
        textfont=dict(size=10, color='#2c3e50'),
        hovertemplate=(
            '<b>⭐ %{text}</b><br>' +
            'Rapport P/P: %{x:.2f}<br>' +
            'Note Presse: %{y:.1f}/10<br>' +
            'Prime Prix: %{marker.color:.1f}%<br>' +
            '<extra>ZONE ELITE</extra>'
        )
    ))

# Lignes de référence pour les quadrants
fig_bubble.add_vline(x=2.0, line_dash="dash", line_color="#e74c3c", line_width=2,
                     annotation_text="P/P = 2.0", annotation_position="top")
fig_bubble.add_hline(y=9.5, line_dash="dash", line_color="#9b59b6", line_width=2,
                     annotation_text="Note = 9.5", annotation_position="right")

# Annotations des zones
fig_bubble.add_annotation(
    x=1.0, y=10.0,
    text="🏆 ZONE ELITE<br>(P/P < 2.0 + Note ≥ 9.5)",
    showarrow=False,
    font=dict(size=12, color='#27ae60', family='Arial Black'),
    bgcolor='rgba(255,255,255,0.9)',
    bordercolor='#2ecc71',
    borderwidth=2,
    borderpad=6
)

fig_bubble.add_annotation(
    x=1.25, y=9.25,
    text="⭐ Premium",
    showarrow=False,
    font=dict(size=10, color='#f39c12'),
    bgcolor='rgba(255,255,255,0.8)'
)

fig_bubble.add_annotation(
    x=1.0, y=7.5,
    text="🔧 Performante",
    showarrow=False,
    font=dict(size=10, color='#3498db'),
    bgcolor='rgba(255,255,255,0.8)'
)

fig_bubble.add_annotation(
    x=4.0, y=7.0,
    text="📊 Standard",
    showarrow=False,
    font=dict(size=10, color='#7f8c8d'),
    bgcolor='rgba(255,255,255,0.8)'
)

# Mise en page
fig_bubble.update_layout(
    title=dict(
        text='🎯 Bubble Chart : Analyse Multi-Critères des Véhicules',
        font=dict(size=18)
    ),
    xaxis=dict(
        title='Rapport Poids/Puissance (kg/ch)',
        range=[0, max(df_bubble['Rapport Poids/Puiss']) * 1.1],
        gridcolor='rgba(0,0,0,0.1)'
    ),
    yaxis=dict(
        title='Note Presse (/10)',
        range=[6, 10.5],
        gridcolor='rgba(0,0,0,0.1)'
    ),
    template='plotly_white',
    legend=dict(
        x=0.02, y=0.02,
        bgcolor='rgba(255,255,255,0.9)',
        bordercolor='#ddd',
        borderwidth=1
    ),
    height=600
)

fig_bubble.show()

# Tableau récapitulatif
print("\n" + "=" * 80)
print("🎯 ANALYSE DES ZONES DE POSITIONNEMENT")
print("=" * 80)

# Statistiques par zone
zones = ['Elite', 'Standard']
for zone in zones:
    df_zone = df_bubble[df_bubble['Zone'] == zone]
    if len(df_zone) > 0:
        emoji = "🏆" if zone == 'Elite' else "📊"
        print(f"\n{emoji} ZONE {zone.upper()} ({len(df_zone)} véhicules):")
        print("-" * 60)
        print(f"   • Rapport P/P moyen: {df_zone['Rapport Poids/Puiss'].mean():.2f} kg/ch")
        print(f"   • Note Presse moyenne: {df_zone['Note Presse (/10)'].mean():.2f}/10")
        print(f"   • Prime de Prix moyenne: {df_zone['Prime de Prix (%)'].mean():+.1f}%")
        print(f"   • Popularité moyenne: {df_zone['Score Popularité'].mean():.1f}/100")

# Liste des véhicules Elite
if len(df_elite) > 0:
    print("\n" + "=" * 80)
    print("🏆 VÉHICULES ELITE (P/P < 2.0 + Note ≥ 9.5)")
    print("=" * 80)
    print(f"{'Marque':<15} {'Modèle':<25} {'P/P':<8} {'Note':<8} {'Prime':<10}")
    print("-" * 80)
    for _, row in df_elite.sort_values('Prime de Prix (%)', ascending=False).iterrows():
        print(f"{row['Marque']:<15} {str(row['Modèle / Variante'])[:24]:<25} {row['Rapport Poids/Puiss']:.2f}    {row['Note Presse (/10)']:.1f}     {row['Prime de Prix (%)']:+.1f}%")
else:
    print("\n⚠️ Aucun véhicule dans la zone Elite avec les critères actuels")
    print("   (P/P < 2.0 ET Note Presse ≥ 9.5)")

# Recommandations
print("\n" + "=" * 80)
print("💡 RECOMMANDATIONS D'INVESTISSEMENT:")
print("-" * 80)
print("   🏆 Zone Elite: Véhicules rares à forte valeur - investissement prioritaire")
print("   ⭐ Zone Premium: Bon compromis performance/valeur - stock régulier")
print("   🔧 Zone Performante: Attrait sportif mais note moyenne - rotation rapide")
print("   📊 Zone Standard: Volume de marché classique - marge réduite")
```

```python
# Grouped Bar Chart : Comparatif Elite vs Standard

import plotly.graph_objects as go
import numpy as np

# Utiliser les données du Bubble Chart (df_bubble avec colonnes Zone, Prime de Prix, Score Popularité)
# Créer le Délai de vente simulé basé sur la popularité et la zone
np.random.seed(42)

# Ajouter le délai de vente au df_bubble
df_bubble['Délai Vente (jours)'] = df_bubble.apply(
    lambda row: np.random.uniform(15, 35) if row['Zone'] == 'Elite' 
    else np.random.uniform(45, 90), axis=1
)

# Calculer les statistiques par zone
stats_comparatif = []
for zone in ['Elite', 'Standard']:
    df_zone = df_bubble[df_bubble['Zone'] == zone]
    if len(df_zone) > 0:
        stats_comparatif.append({
            'Zone': zone,
            'Prime Moyenne (%)': df_zone['Prime de Prix (%)'].mean(),
            'Délai Vente (jours)': df_zone['Délai Vente (jours)'].mean(),
            'Score Popularité': df_zone['Score Popularité'].mean(),
            'N': len(df_zone)
        })

df_comparatif = pd.DataFrame(stats_comparatif)

# Créer le graphique en barres groupées
fig_comparatif = go.Figure()

# Couleurs par zone
couleurs_zone = {'Elite': '#27ae60', 'Standard': '#95a5a6'}

# Métriques à comparer
metriques = ['Prime Moyenne (%)', 'Délai Vente (jours)', 'Score Popularité']
labels_metriques = ['💰 Prime Prix (%)', '⏱️ Délai Vente (j)', '⭐ Popularité']

# Normaliser pour affichage (échelles différentes)
# On affiche les valeurs brutes mais on note l'avantage Elite
for i, (metrique, label) in enumerate(zip(metriques, labels_metriques)):
    valeurs = df_comparatif[metrique].tolist()
    zones = df_comparatif['Zone'].tolist()
    
    # Couleurs selon la zone
    colors = [couleurs_zone[z] for z in zones]
    
    fig_comparatif.add_trace(go.Bar(
        name=label,
        x=zones,
        y=valeurs,
        marker_color=colors if i == 0 else None,  # Première métrique avec couleurs zones
        text=[f"{v:.1f}" for v in valeurs],
        textposition='outside',
        textfont=dict(size=12),
        offsetgroup=i,
        legendgroup=label,
        showlegend=True if i == 0 else False
    ))

# Recréer le graphique avec une approche différente : métriques en X, zones côte à côte
fig_comparatif = go.Figure()

# Barres pour Elite
elite_data = df_comparatif[df_comparatif['Zone'] == 'Elite'].iloc[0] if len(df_comparatif[df_comparatif['Zone'] == 'Elite']) > 0 else None
standard_data = df_comparatif[df_comparatif['Zone'] == 'Standard'].iloc[0] if len(df_comparatif[df_comparatif['Zone'] == 'Standard']) > 0 else None

if elite_data is not None and standard_data is not None:
    # Normaliser les valeurs pour comparaison visuelle
    # Prime: plus haut = mieux (positif)
    # Délai: plus bas = mieux (inverser l'échelle)
    # Popularité: plus haut = mieux
    
    metriques_labels = ['Prime Prix (%)', 'Délai Vente (j)', 'Popularité']
    
    elite_values = [
        elite_data['Prime Moyenne (%)'],
        elite_data['Délai Vente (jours)'],
        elite_data['Score Popularité']
    ]
    
    standard_values = [
        standard_data['Prime Moyenne (%)'],
        standard_data['Délai Vente (jours)'],
        standard_data['Score Popularité']
    ]
    
    # Barres Elite
    fig_comparatif.add_trace(go.Bar(
        name='🏆 Elite',
        x=metriques_labels,
        y=elite_values,
        marker_color='#27ae60',
        text=[f"{v:.1f}" for v in elite_values],
        textposition='outside',
        textfont=dict(size=13, color='#27ae60')
    ))
    
    # Barres Standard
    fig_comparatif.add_trace(go.Bar(
        name='📊 Standard',
        x=metriques_labels,
        y=standard_values,
        marker_color='#95a5a6',
        text=[f"{v:.1f}" for v in standard_values],
        textposition='outside',
        textfont=dict(size=13, color='#7f8c8d')
    ))
    
    # Ajouter des annotations pour montrer l'avantage Elite
    avantages = []
    for i, (e, s, label) in enumerate(zip(elite_values, standard_values, metriques_labels)):
        if 'Délai' in label:
            # Pour délai, moins c'est mieux
            diff = ((s - e) / s) * 100  # % de réduction
            avantages.append(f"-{abs(diff):.0f}%")
        else:
            # Pour les autres, plus c'est mieux
            diff = ((e - s) / abs(s)) * 100 if s != 0 else 0
            avantages.append(f"+{diff:.0f}%" if diff > 0 else f"{diff:.0f}%")
    
    # Annotations d'avantage
    for i, (label, avantage) in enumerate(zip(metriques_labels, avantages)):
        max_val = max(elite_values[i], standard_values[i])
        fig_comparatif.add_annotation(
            x=label,
            y=max_val * 1.15,
            text=f"<b>{avantage}</b>",
            showarrow=False,
            font=dict(size=14, color='#27ae60' if '+' in avantage or '-' in avantage else '#e74c3c'),
            bgcolor='rgba(255,255,255,0.9)',
            bordercolor='#27ae60',
            borderwidth=1,
            borderpad=3
        )

# Mise en page
max_y = max(max(elite_values), max(standard_values)) if elite_data is not None else 100
fig_comparatif.update_layout(
    title=dict(
        text='📊 Comparatif : Véhicules Elite vs Standard',
        font=dict(size=18)
    ),
    xaxis_title='Métrique',
    yaxis_title='Valeur',
    template='plotly_white',
    barmode='group',
    bargap=0.25,
    bargroupgap=0.1,
    legend=dict(
        x=0.5, y=1.12,
        xanchor='center',
        orientation='h',
        bgcolor='rgba(255,255,255,0.9)'
    ),
    yaxis=dict(range=[0, max_y * 1.35])
)

# Ajouter une annotation résumé
fig_comparatif.add_annotation(
    x=0.5, y=-0.15,
    xref='paper', yref='paper',
    text="✅ Les véhicules Elite cumulent : Prime plus élevée + Délai réduit + Popularité supérieure",
    showarrow=False,
    font=dict(size=12, color='#27ae60'),
    bgcolor='rgba(39, 174, 96, 0.1)',
    bordercolor='#27ae60',
    borderwidth=1,
    borderpad=8
)

fig_comparatif.show()

# Tableau récapitulatif
print("\n" + "=" * 80)
print("📊 COMPARATIF ELITE vs STANDARD")
print("=" * 80)
print(f"{'Métrique':<25} {'🏆 Elite':<20} {'📊 Standard':<20} {'Avantage Elite':<15}")
print("-" * 80)

if elite_data is not None and standard_data is not None:
    # Prime de prix
    e_prime = elite_data['Prime Moyenne (%)']
    s_prime = standard_data['Prime Moyenne (%)']
    diff_prime = e_prime - s_prime
    print(f"{'Prime Prix (%)':<25} {e_prime:>+15.1f}%    {s_prime:>+15.1f}%    {diff_prime:>+10.1f}%")
    
    # Délai de vente
    e_delai = elite_data['Délai Vente (jours)']
    s_delai = standard_data['Délai Vente (jours)']
    diff_delai = s_delai - e_delai  # Inverser car moins = mieux
    pct_delai = (diff_delai / s_delai) * 100
    print(f"{'Délai Vente (jours)':<25} {e_delai:>15.1f}j     {s_delai:>15.1f}j     {pct_delai:>+9.0f}% ⏱️")
    
    # Popularité
    e_pop = elite_data['Score Popularité']
    s_pop = standard_data['Score Popularité']
    diff_pop = e_pop - s_pop
    print(f"{'Score Popularité':<25} {e_pop:>15.1f}      {s_pop:>15.1f}      {diff_pop:>+10.1f} pts")
    
    # Nombre de véhicules
    n_elite = int(elite_data['N'])
    n_standard = int(standard_data['N'])
    print(f"\n{'Nombre de véhicules':<25} {n_elite:>15}       {n_standard:>15}")

print("=" * 80)

# Conclusion
print("\n🏆 CONCLUSION:")
print("-" * 80)
print("   Les véhicules ELITE cumulent 3 avantages majeurs :")
print(f"   1. 💰 Prime de prix supérieure : +{diff_prime:.1f}% vs Standard")
print(f"   2. ⏱️ Délai de vente réduit : -{pct_delai:.0f}% (vente plus rapide)")
print(f"   3. ⭐ Popularité plus élevée : +{diff_pop:.1f} points")
print("\n   → Stratégie recommandée : Prioriser l'acquisition de véhicules Elite")
print("   → ROI optimal grâce à la combinaison marge élevée + rotation rapide")
```
